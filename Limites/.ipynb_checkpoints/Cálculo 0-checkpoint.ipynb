**Definição freestyle :**

O limite de uma função em $p$ é como ela *se comporta* (pra onde ela vai) conforme $x$ chega mais perto de $p$.

**Obs: o que acontece exatamente em $p$ NÃO IMPORTA**

<br>


<div class="jumbotron">
Dizer que o limite de $f(x)$ quando $x$ se aproxima de $p$ é igual à $L$ equivale, simbólicamente, à

$$
\lim_{x \rightarrow p}f(x) = L
$$

</div>


Também vale lembrar que 

$$
\lim_{x \rightarrow p}f(x) \; existe \; \Leftrightarrow \; \lim_{x \rightarrow +p}f(x) = lim_{x \rightarrow -p}f(x)
$$ 

Isto é, não importa se estamos nos aproximando pela direita ou esquerda, o resultado é o mesmo.

<br> 

### Exemplo 0.1 :
- **Tome $f:\mathbb{R} \rightarrow \mathbb{R},$**
$$
f(x) =
       \begin{cases}
             0, & \mbox{se $x < 1$} \\
             1, & \mbox{se $x \geq 1$}
       \end{cases}
$$ 

temos que 
$$
\lim_{x \rightarrow +1}f(x) = 1
$$
e
$$
\lim_{x \rightarrow -1}f(x) = 0
$$

Desta forma, $\lim_{x \rightarrow 1}f(x)$ *não existe*.

### Exemplo 0.2 : 
- **Qual o limite de $f:\mathbb{R} \rightarrow \mathbb{R},\;f(x) = x + 1$ quando $x$ se aproxima de $1$??**

Vamos ver o gráfico :
![](./img/img1.png)

**R**: Podemos ver que conforme $x$ vai chegando perto de $1$, $f(x)$ tende à $2$.


Neste caso acontece que $\lim_{x \rightarrow 2} f(x) = f(2)$, mas agora vamos ver outro caso.

### Exemplo 0.3 :
- **Qual o limite de $f:\mathbb{R} \rightarrow \mathbb{R}$,**
$$ 
f(x) = 
       \begin{cases}
             0, & \mbox{se $x =1$} \\
             x+1, & \mbox{caso contrário.}
       \end{cases}
$$

**quando $x$ se aproxima de $1$??**

O gráfico fica
![](./img/img2.png)

Repare que ele é quase igual ao último, sendo que a única diferença está em $x=1$. Mas como dissemos anteriormente, isso não importa muito, já que *'o limite só se importa com a vizinhança'*.

Temos então que, assim como no último exemplo, $\lim_{x \rightarrow 1} f(x) = 2$, mesmo que $f(2) = 0$.

<br>

Essa ideia também pode ser usada em casos em que a função apresenta algum "problema":

### Exemplo 0.4 : 

- **Qual o limite de $f(x) = \frac{x^{2} - 1}{x - 1}$ quando $x$ se aproxima de $1$??**

Podemos notar que essa função tem uma indeterminação em $x = 1$ (se tentarmos substituir diretamente em $f$ acabamos com $f(1) = \frac{1 - 1}{1 - 1} = \frac{0}{0}$, o que é vai a constituição brasileira). Mas podemos fazer algumas manipulações com ela:

$$
f(x) = \frac{x^2 - 1}{x - 1} = \frac{(x-1)(x+1)}{(x-1)}
$$

Mas antes de cortar o denominador, lembre que a função ainda tem esse problema. Um jeito de resolvê-lo seria fazer:

$$
f(x) = 
       \begin{cases}
             x+1, & \mbox{se $x \neq 1$} \\
             \mbox{???}, & \mbox{caso contrário.}
       \end{cases}
$$

O segundo caso não nos interessa tanto, como vimos no último exemplo. Mas olhando para a vizinhança podemos notar que, novamente, $\lim_{x \rightarrow 1} f(x) = 2$.

<br>

**Limites infinitos :**

No último exemplo vimos o caso de função com uma indeterminação em um ponto mas que o limite nele era **definido**. Entretanto também temos casos em que a função *diverge*, isto é, vai para o "infinito positivo ou negativo".

### Exemplo 0.5 :

- Vamos pegar a função $f:\mathbb{R} \setminus \{1\} \rightarrow \mathbb{R}, \; f(x) = \frac{1}{(x-1)^2}$. Seu gráfico fica:

![](./img/img4.png)

Assim como no último caso, temos um problema em $x=1$, mas agora ele também se estende à sua vizinhança. Seguindo a lógica de para onde vai $f(x)$ conforme $x$ se aproxima de $p=1$, temos que 
$$
\lim_{x \rightarrow 1} \frac{1}{(x-1)^2} = \infty
$$

<br>

**Limites no infinito :**

Não confunda com o último caso! Limites **no** infinito são aqueles em que o ponto $p$ está no infinito (seja positivo ou negativo). Pode ser meio difícil pegar a intuição disso, mas vamos tentar enxergar:

### Exemplo 0.6 :

- Tomemos as funções 

$$
f:\mathbb{R} \setminus \{1\} \rightarrow \mathbb{R}, \; f(x) = \frac{1}{(x-1)^2},
$$

$$
h:\mathbb{R} \rightarrow \mathbb{R}, \; h(x) = x,
$$ 

$$g:\mathbb{R} \rightarrow \mathbb{R}, \; g(x) = \sin(x)
$$ 

![](./img/img3.png)

O caso de $h(x)$ é fácil de ver que vai para o infinito, mas as outras são casos interessantes.


Para $f$ temos que 
$$
\lim_{x \rightarrow \infty} \frac{1}{(x-1)^2} = \frac{1}{\lim_{x \rightarrow \infty} (x-1)^2} = \frac{1}{\infty} = 0
$$
(veremos melhor essa regra daqui a pouco)


Por último, temos $\lim_{x \rightarrow \infty} \sin(x)$. Esse caso é um pouco mais interessante... Note que a função $\sin(x)$ oscila em torno de $0$, mas não converge para valor nenhum. Temos então que $\lim_{x \rightarrow \infty} \sin(x)$ **não existe**.

## 1. Continuidade

</div>

Agora vamos ver um conceito que vai nos seguir pelo resto da graduação (além de todos os outros em cálculo 1): o de continuídade.

A grosso modo, uma função contínua é uma em que você não tira a caneta do papel para desenhar, e no caso de uma função descontínua, cada descontinuidade é um lugar em que você precisa levantar a caneta.

### Exemplo 1.1 :

A função $f:\mathbb{R} \rightarrow \mathbb{R}, \; f(x) = \cos(x)$ é **contínua**
![](./img/img5.png)

A função $g:\mathbb{R} \rightarrow \mathbb{R},$

$$
g(x) = 
       \begin{cases}
             1, & \mbox{se $x > 0$} \\
             -1, & \mbox{se $x < 0$} \\
             0, & \mbox{se $x = 0$}
       \end{cases}
$$

é **descontínua** em $0$.
![](./img/img6.png)


**Obs :** uma função ser contínua ou não *depende do seu domínio*! Uma função pode ter descontinuidades em $\mathbb{R}$ mas conínua em um subconjunto $A \subset \mathbb{R}$. Além disso, **não faz sentido** falarmos continuidade de $f$ em um ponto $p \notin D(f)$ (que não está em seu domínio).

<br>

Vamos tentar colocar essa idéia em linguagem matemágica.

####  Definição
Uma função é dita contínua em $p \in D(f)$ se e somente se para todo $\epsilon > 0$ existe algum $\delta > 0$ tal que, para todo $x \in D(f)$,

$$
|x-p| < \delta \; \Rightarrow \; |f(x) - f(p)| < \epsilon
$$

Se $f$ é contínua em todo $p \in D(f)$ dizemos que $f$ é contínua em $D(f)$. 

<br>

Um modo de verificar a contínuidade (além de diretamente pela definição), é o seguinte:

<div class="jumbotron">

#### Teorema

Tome $f:D(f) \rightarrow \mathbb{R}$ una função, temos que 

$$
    \mbox{$f$ contínua em $p$} \Leftrightarrow \lim_{x \rightarrow p}f(x) = f(p)
$$

Isto é, se $lim_{x \rightarrow p}f(x)$ existe (os limites laterais são iguais) e é igual à $f(p)$, então $f$ é conínua em $p$, e o contrário também vale.

## 2. Teorema do Confronto

</div>
<br>
<img src="./img/jojo.png" style="width: 30em; float: left">

Sejam $f,\;g,\;h$ funções e suponhamos que exista $r > 0$ tal que
$$
g(x) \leq f(x) \leq h(x)
$$
para $|x-p|<r$. Nestas condições, se

$$
\lim_{x \rightarrow p} g(x) = L = \lim_{x \rightarrow p} h(x)
$$

então

$$
\lim_{x \rightarrow p} f(x) = L
$$


Talvez não esteja muito claro agora a importância desse teorema, mas para já ilustrar o tamanho do seu poder, digamos que ele costuma ser usado quando você tem uma $f(x)$ muito complicada de se calcular o limite diretamente, mas você conhece outras duas funções $f$ e $h$ menor e maior que ela que são fáceis de se calcular no ponto desejado. Daqui a pouco vamos ver alguns exemplos.

### 3.1 O Limite Fundamental $\lim_{x \rightarrow 0} \frac{\sin(x)}{x}$

</div>    

Essa é a famosa aproximação do engenheiro: $\sin(\theta) \approx \theta$ para pequenos ângulos. Note que isso é equivalente a dizer que $\frac{\sin(\theta)}{\theta} \approx 1$, ou seja, para $\theta$ muito pequeno isso é **quase** verdade, entretanto ainda tem um pequeno erro (NÃO PERCAM A AULA DE POLINÔMIO DE TAYLOR). Mas para o limite em que $\theta$ vai para zero, temos que.

$$
\lim_{\theta \rightarrow 0}\frac{\sin(\theta)}{\theta} = 1
$$

Essa aproximação das funções pode ser inclusive vista gráficamente:

![](./img/img7.png)

Temos algumas formas de provar isso, mas não vamos nos aprofundar muito nessa parte, mas só para dar uma pequena ilustrada, geométricamente podemos provar que existe $r>0$ tal que

$$
0< \sin(x) < x < \tan(x)
$$

![](./img/img8.png)

E se dividirmos tudo por $\sin(x)$:

$$
1 < \frac{x}{\sin(x)} < \frac{1}{\cos(x)}
$$

Para $0<x<r$,

$$
\cos(x) < \frac{\sin(x)}{x} < 1
$$

Agora temos uma condição perfeita para usar o teorema do confronto, já que

$$
\lim_{x \rightarrow 0}\cos(x) = \lim_{x \rightarrow 0} 1 = 1
$$

$$
\Rightarrow \lim_{x \rightarrow 0}\frac{\sin(x)}{x} = 1
$$

### 3.2 Limite de Função Composta 

</div>
    
Lembrando da definição de uma função composta, é básicamente uma "função de outra função". Imagina que eu tenho por exemplo $f(t) = \sqrt{t}$ e $g(t) = t - 4$. A função composta $f(g(t))$ ou $(f \circ g)(t)$ é dada por:

$$
f(g(t)) = \sqrt{t - 4 }
$$

<br>

Trabalhando com limites, vamos ter que:

$$
\lim_{x \rightarrow p}f(g(x)) = \lim_{u \rightarrow a} f(u) 
$$

contanto que $\lim_{u \rightarrow a}f(u)$ exista. Note que $u = g(x); \; u \rightarrow a \; para \;  x \rightarrow p$

Pode ser meio complicado de visualizar o que está acontecendo nesse formato, mas vamos tentar ilustrar:

### Exemplo 3.1 :
- **Calcule**
$$
\lim_{x \rightarrow 0} \frac{\sin(5x)}{2x}
$$


Podemos tratar isso como uma composição de funções que sabemos resolver! Tome $y(x) = 5x$, de forma que 

$$
\lim_{x \rightarrow 0} y(x) = 0
$$

Então, sabendo que $x = \frac{y}{5}$, simplesmente substituo na função original :

$$
\frac{\sin(5x}{2x} = \frac{\sin(y)}{\frac{2}{5}y} = \frac{5}{2}\frac{\sin(y)}{y}
$$

Como conforme $x \rightarrow 0$, $y$ também vai para zero, temos que

$$
\lim_{x \rightarrow 0} \frac{\sin(5x)}{2x} = \lim_{y \rightarrow 0} \frac{5}{2}\frac{\sin(y)}{y} =
$$

$$
\frac{5}{2} \lim_{y \rightarrow 0} \frac{\sin(y)}{y}
$$

O problema se transforma no nosso conhecido limite fundamental, assim nos deixando com:

$$
\lim_{x \rightarrow 0} \frac{\sin(5x)}{2x} =  \frac{5}{2}
$$

### 3.3 Propriedades operatórias
    
</div>

Vamos discutir algumas propriedades dos limites que podem ajudar em algumas situações. Elas podem parecer meio óbvias mas a matemática é assim mesmo às vezes

<br>

<div class="jumbotron">

Tome $f$ e $g$ funções tal que $\lim_{x \rightarrow p} f(x) = L_1$ e $\lim_{x \rightarrow p} g(x) = L_2$ e um número qualquer $k$ constante. Temos que: 

1. $ \lim_{x \rightarrow p} (f(x) + g(x)) = \lim_{x \rightarrow p} f(x) + \lim_{x \rightarrow p} g(x) = L_1 + L_2$
   
<br>
   
2. $\lim_{x \rightarrow p} kf(x) = k\lim_{x \rightarrow p} f(x) = kL_1$
   
<br>
   
3. $\lim_{x \rightarrow p} f(x)g(x) = \lim_{x \rightarrow p} f(x)\lim_{x \rightarrow p} g(x)$

<br>

4. $\lim_{x \rightarrow p} \frac{f(x)}{g(x)} = \frac{\lim_{x \rightarrow p} f(x)}{\lim_{x \rightarrow p} g(x)} = \frac{L_1}{L_2}$, se $L_2 \neq 0$

</div>
    
Todas essas propriedades podem ser tiradas diretamente da definição de limite (com os $\epsilon$  e $\delta$).

### 3.4 Algumas dicas para os exercícios

Agora já vimos a maior parte das propriedades "importantes" dos limites, oque significa que teóricamente vocês já tem as ferramentas necessárias para resolver boa parte dos exercícios que possam aparecer. Deixei essa seção reservada aqui no final para, de um modo bem menos formal e mais prático, tratar os conceitos mais usados para a resolução de problemas e alguns macetes.


Normalmente podemos dividir os problemas de limites em certos tipos mais simples:

<br>

#### Aqueles em que $f(x)$ é conínua em $p$ então $\lim_{x \rightarrow p} f(x) = f(p)$ 

São os mais simples, já que podemos simplesmente substituir o valor de $p$ na função.

**Ex :** $\lim_{x \rightarrow 0} 4x = 0$ ou $\lim_{x \rightarrow 10} 4x = 40$

<br>

#### Aqueles em que ocorre $\lim_{x \rightarrow p} f(x) =\infty$

Não tem o que fazer, o limite vai para o infinito mesmo.

**Ex :** $\lim_{x \rightarrow +\infty} 4x = +\infty$

<br>

#### Aqueles em que ficamos com $\frac{c}{0}$, $c$ constante

Se nos deparamos por exemplo com alguma constante sobre algo que vai para zero, isso sempre diverge para $\infty$

**Ex :** $\lim_{x \rightarrow 0} \frac{19}{x^2} = \infty$

<br>

#### Aqueles em que ficamos com $\frac{c}{\infty}$, $c$ constante

Da mesma forma, um número constante sobre algo que fica cada vez maior tende para zero.

**Ex :** $\lim_{x \rightarrow +\infty} \frac{19}{(x^4 + 3x)} = 0$

<br>


#### Aqueles em que ficamos com $\frac{\infty}{\infty}$ ou $\frac{0}{0}$

Esses são os mais interessantes. A princípio **não podemos dizer nada** sobre eles. Podem resultar tanto em $0$, quanto em algum outro número ou até divergirem. Isso depende de *"quão rápido"* o denominador e o númerador vão para zero ou para o infinito. Normalmente essas são os casos que nos dão problema.

Existem diversas formas de resolvê-los, como vimos durante a aula. Um caso mais simples é o de *polinômios*, especialmente aqueles em que a variável vai para zero ou infinito!.

**Ex :** Considere $\lim_{x \rightarrow \infty}\frac{x^3 -4x}{x^8 + 2x^4 + 2x -1}$.

Ele se enquadra no caso $\frac{\infty}{\infty}$. Podemos resolvê-lo fácilmente da seguinte forma: primeiro divido tudo pela maior potência do numerador:

$$
\lim_{x \rightarrow \infty}\frac{x^3 -4x}{x^8 + 2x^4 + 2x -1} = \lim_{x \rightarrow \infty}\frac{\frac{x^3}{x^3} - \frac{4x}{x^3}}{\frac{x^8}{x^3} + \frac{2x^4}{x^3} + \frac{2x}{x^3} - \frac{1}{x^3}} =
$$

$$
= \lim_{x \rightarrow \infty} \frac{1 - \frac{1}{x^2}}{x^5 + 2x + \frac{2}{x^2} - \frac{1}{x^3}}
$$

E usando as propriedades que vimos, podemos notar que caímos para um caso $\frac{constante}{\infty}$, o que resulta em 

$$
\lim_{x \rightarrow \infty}\frac{x^3 -4x}{x^8 + 2x^4 + 2x -1} = 0
$$

Também podemos fazer algo parecido para o caso em que $x \rightarrow 0$. De forma simples, posso dizer que **o termo com a maior potência prevalece**. Isso tem a ver com o que foi comentado sobre o "quão rápido" ele se aproxima de $0$ ou $\infty$.

Um outro caso interessante acontece quando as potências são iguais.

**Ex :** Considere $\lim_{x \rightarrow \infty}\frac{5x^8 + x^3 -4x}{2x^8 + 2x^4 + 2x -1}$.

Se usarmos o mesmo método, podemos ver que ficaremos com

$$
\lim_{x \rightarrow \infty}\frac{5 + \frac{1}{x^5}-\frac{4}{x^7}}{2 + \frac{2}{x^4} + \frac{2}{x^7} - \frac{1}{x^8}} = 
$$

$$
 = \frac{5}{2}
$$


Essa propriedade pode ser usada com qualquer polinômio, embora alguns precisem de alguma manipulação (por composição de funções, por exemplo) e nem todos sejam complicados a ponto de precisarem disso, muitas vezes é melhor ficarmos só na manipulação algébrica.
